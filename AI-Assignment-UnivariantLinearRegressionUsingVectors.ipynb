```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv("Data-Files/HousePricesData.csv")
```

```python
dataset
```

```python
x_size = dataset["House Size (sq meters)"].tolist()
print(x_size)
```

```python
y_price = dataset["Price (USD)"].tolist()
print(y_price)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x_size,y_price,marker='x',color='r',label='Data Points')
plt.xlabel("Size of the house in square meters")
plt.ylabel("Price of the house in USD $")
plt.title("House Sizes Data with thier prices")
plt.legend()
plt.show()
```

```python
x = np.array(x_size).reshape(-1,1)
y = np.array(y_price).reshape(-1,1)
m = len(y)
```

```python
x = np.column_stack((np.ones((m, 1)), x))
```

```python
def hypothesis(theta, X):
    return np.dot(X, theta)

def compute_cost(theta, X, y):
    m = len(y)
    predictions = hypothesis(theta, X)
    errors = predictions - y
    cost = (1 / (2 * m)) * np.dot(errors.T, errors)
    return cost

def compute_gradient(theta, X, y):
    m = len(y)
    predictions = hypothesis(theta, X)
    errors = predictions - y
    gradient = (1 / m) * np.dot(X.T, errors)
    return gradient
```

```python
w = np.zeros((2,1))
epochs = []
costs = []
for i in range(15):
    epochs.append(i)
    cost = compute_cost(w,x,y)
    costs.append(cost[0][0])
    gradient = compute_gradient(w, x, y)
    w -= 0.01 * gradient
plt.plot(epochs,costs,c='red')
plt.xlabel("Epochs")
plt.ylabel("Costs")
plt.title("Costs Minimization at Learning rate of 0.01")
plt.show()
```

```python
b = w[1][0]
w = w[0][0]
print(w,b)
```

```python
def predict(x):
    return ((w*x) + b) 
```

```python
x_values = [i*2 for i in range(1,200)]
predictions = []
for size in x_values:
    predictions.append(predict(size)) 
plt.plot(x_values,predictions,c='g',label='Fine Fit')
plt.xlabel("Random Sizes of the Houses")
plt.ylabel("Predictions (Prices of Houses) of the Model")
plt.legend()
plt.show()
```
